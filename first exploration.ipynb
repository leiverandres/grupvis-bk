```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

CONV_RESULTS_PATH = "./datasets/resultados-preliminares-grupos-conv_781_L.csv"
```

```python
data = pd.read_csv(CONV_RESULTS_PATH)
data.info()
```

```python
data.head()
```

```python
data["INSTITUCIONES QUE LO AVALAN"].value_counts()
```

```python
# Remove newlines
data["INSTITUCIONES QUE LO AVALAN"] = data["INSTITUCIONES QUE LO AVALAN"].apply(lambda x: ''.join(str(x).split('\n|-')))
# Substitute multiple whitespace with single whitespace
data["INSTITUCIONES QUE LO AVALAN"] = data["INSTITUCIONES QUE LO AVALAN"].apply(lambda x: ' '.join(str(x).split()))
data["INSTITUCIONES QUE LO AVALAN"].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
data["INSTITUCIONES QUE LO AVALAN"].value_counts()[:20].plot(kind="bar", ax=ax);
```

```python
data.to_csv('cleaned_up.csv', index=False)
```
